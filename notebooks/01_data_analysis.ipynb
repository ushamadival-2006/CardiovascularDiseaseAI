# 01_data_analysis.ipynb

This notebook performs initial data loading and exploratory data analysis (EDA) for the Cardiovascular Disease dataset (cardio_train.csv).
Place `cardio_train.csv` in the `data/` folder (one level above this `notebooks/` folder) so the notebook can load it using the relative path `../data/cardio_train.csv`.

```python
# Import required libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Set visual style
sns.set(style="whitegrid")
```

```python
# Define dataset path (expected in data/cardio_train.csv relative to repo root)
dataset_path = os.path.join('..', 'data', 'cardio_train.csv')

# Load the dataset with a friendly error message if the file is missing
try:
    df = pd.read_csv(dataset_path)
    print(f'Loaded dataset from: {dataset_path}')
except FileNotFoundError:
    raise FileNotFoundError(f'Could not find cardio_train.csv at {dataset_path}. Please place the file in the data/ folder.')
```

```python
# 2. Display the first 5 rows to inspect the data
df.head()
```

```python
# 3. Print dataset shape (rows, columns)
print('Dataset shape:', df.shape)
```

```python
# 4. Print all column names
print('Columns:', list(df.columns))
```

```python
# 5. Display data types of all columns
df.dtypes
```

```python
# 6. Check for missing values in every column
missing = df.isnull().sum()
print('Missing values per column:\n', missing)
```

```python
# 7. Check for duplicate records
num_duplicates = df.duplicated().sum()
print('Number of duplicate rows:', num_duplicates)
```

```python
# 8. Print basic statistics using describe()
df.describe(include='all').transpose()
```

## Target variable distribution
We will examine the distribution of the `cardio` target variable (0 = no disease, 1 = disease).

```python
# 9. Show the distribution of the target variable `cardio`
plt.figure(figsize=(6,4))
sns.countplot(x='cardio', data=df, hue='cardio', palette='pastel', legend=False)
plt.title('Heart Disease (cardio) Distribution')
plt.xlabel('cardio')
plt.ylabel('Count')
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().annotate(format(int(height), ',d'), (p.get_x()+p.get_width()/2., height), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()

# Also print value counts for clarity
print(df['cardio'].value_counts())
```

## Age distribution
Many cardiovascular datasets store `age` as days; we convert to years for readability.

```python
# 10. Age distribution (convert days -> years if needed)
if 'age' in df.columns:
    # Create a new column in years (rounded) for easier interpretation
    df['age_years'] = (df['age'] / 365).astype(int)
    plt.figure(figsize=(8,4))
    sns.histplot(df['age_years'], bins=30, kde=True, color='skyblue')
    plt.title('Age Distribution (years)')
    plt.xlabel('Age (years)')
    plt.ylabel('Count')
    plt.show()
    print(df['age_years'].describe())
else:
    print('No `age` column found in dataset.')
```

## Blood pressure distribution
We will plot systolic (`ap_hi`) and diastolic (`ap_lo`) distributions if present.

```python
# 11. Blood pressure distributions
fig, axes = plt.subplots(1, 2, figsize=(14,4))
if 'ap_hi' in df.columns:
    sns.histplot(df['ap_hi'], bins=30, kde=True, ax=axes[0], color='salmon')
    axes[0].set_title('Systolic Blood Pressure (ap_hi)')
    axes[0].set_xlabel('ap_hi')
else:
    axes[0].text(0.5, 0.5, 'ap_hi not found', ha='center', va='center')
if 'ap_lo' in df.columns:
    sns.histplot(df['ap_lo'], bins=30, kde=True, ax=axes[1], color='lightgreen')
    axes[1].set_title('Diastolic Blood Pressure (ap_lo)')
    axes[1].set_xlabel('ap_lo')
else:
    axes[1].text(0.5, 0.5, 'ap_lo not found', ha='center', va='center')
plt.tight_layout()
plt.show()
```

## Cholesterol distribution
Plot cholesterol categories or values if the column exists.

```python
# 12. Cholesterol distribution
if 'cholesterol' in df.columns:
    plt.figure(figsize=(6,4))
    # If cholesterol is categorical (1,2,3), countplot is useful
    try:
        sns.countplot(x='cholesterol', data=df, hue='cholesterol', palette='muted', legend=False)
        plt.title('Cholesterol Distribution (categories)')
        plt.xlabel('cholesterol')
        plt.ylabel('Count')
        plt.show()
    except Exception:
        # Fall back to histogram for numeric values
        sns.histplot(df['cholesterol'].dropna(), bins=30, kde=True, color='violet')
        plt.title('Cholesterol Distribution')
        plt.xlabel('cholesterol')
        plt.ylabel('Count')
        plt.show()
else:
    print('No `cholesterol` column found in dataset.')
```
